```python
import pandas as pd


prices = pd.read_csv('prices.csv')
sales = pd.read_csv('sales.csv')
print(prices)
```

```python
import numpy as np
sales = sales.sort_values(by='ordered_at')
sales['row_num'] = np.arange(len(sales))
sales
```

```python
#convert ordered_at, updated_at to datetime object (using pd.to_datetime)
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
prices['updated_at'] = pd.to_datetime(prices['updated_at'])
```

```python
prices = prices.sort_values(by='updated_at')
prices
```

```python
df = sales.merge(prices, left_on='product_id', right_on='product_id')
df['ordered_updated'] = df['ordered_at']-df['updated_at']
df
```

```python
# TH1: lấy những dòng tạo đơn sau cập nhật
import numpy as np  
df1= df[df['ordered_updated'] >= '-0 days +0:00:00']
```

```python
df1['rn'] = df1.groupby(['product_id','ordered_at','row_num'])['ordered_updated'].rank(method='first')
```

```python
df1= df1[df1['rn'] == 1]   
df1['gia_hientai'] = df1['new_price']
df1
```

```python
# TH2: lấy những dòng tạo đơn trước cập nhật (khác df1)

df2= df[df['ordered_updated'] < '-0 days +0:00:00']
df2['rn'] = df2.groupby(['product_id','ordered_at','row_num'])['ordered_updated'].rank(ascending = False)
df2= df2[df2['rn'] == 1]  
df2['new'] = df2['product_id'].astype(str)+df2['ordered_at'].astype(str)+df2['row_num'].astype(str)
```

```python
a= df1['product_id'].astype(str)+df1['ordered_at'].astype(str)+df1['row_num'].astype(str)
df2 = df2[~df2['new'].isin(a)]
```

```python
df2.drop(columns='new',inplace=True)
df2['gia_hientai'] = df2['old_price']
```

```python
df_union= pd.concat([df1, df2]).drop_duplicates()
print(df_union)
```

```python
# Tính doanhthu
df_union['doanhthu'] = df_union['quantity_ordered'] * df_union['gia_hientai']
doanhthu_giahientai= df_union.groupby(['product_id', 'gia_hientai'], as_index=False)['doanhthu'].sum()
doanhthu_giahientai
```

```python
total_revenue = df_union.groupby('product_id')['doanhthu'].sum()
total_revenue
```
